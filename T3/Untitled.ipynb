# WAP to find longest word that uses a,b,c,d,e,f, in any order

```python
longest = ""
with open("wordlist.txt") as source:
    for word in source.readlines():
        if {'a','b','c','d','e','f'}.issubset(set(word)) and len(word.strip()) > len(longest):
            longest = word.strip()
print(longest)
```

# WAP to find the word with maximum 'i'

```python
req=""
with open("wordlist.txt") as file:
    for word in file.readlines():
        if word.count("i")>req.count("i"):
            req=word
print(req)
```

```python
with open("wordlist.txt") as file:
    l = {word.count("i") : word for word in file.read().split()}
    print(l)
    print(l[max(l)])
```

# WAP to extract japanese text from the file 
`secretmassge.txt`
- SECOND TASK:-
    - The meaning of the text is the first word of each line. Extract the meaning

```python
with open("secretmessage.txt",encoding='utf-8') as file:
    for line in file:
        for word in line.split():
            if word[0].isalnum() != True:
                japanese = word[1:len(word) - 1]
        print(line.split()[0], end=" ")
print()
print("word : " , japanese)
```
