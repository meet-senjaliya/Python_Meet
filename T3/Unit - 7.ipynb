````python
def formatlines(inputfile,outputfile,maxlength=80):
    infile = open(inputfile)
    outfile = open(outputfile,'w')
    for line in infile:
        words = line.split()
        current = ""
        for word in words:
            if current == "":
                newline = word
            else:
                newline = current + " " + word
            if len(newline) <= maxlength:
                current = newline
            else:
                outfile.write(current + "\n")
                current = word
        if current:
            outfile.write(current + "\n")
    infile.close()
    outfile.close()
````
