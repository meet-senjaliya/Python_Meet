## WAP to implement the find method in string

````python
def findchar(s,char):
    ind = 0
    for i in s:
        if i==char:
            return ind
        ind += 1   
    else:
        return -1
````

## WAP a programm to shift a number based on following condition
```
input - 12345

shift - 1 -> o/p = 23451

shift - 2 -> o/p = 34512

shift - 3 -> o/p = 45123

shift - 4 -> o/p = 51234

shift - 5 -> o/p = 12345

shift - 6 -> o/p = 54321

````python
def shifting(number,shift):
    strnumber=(str(number))
    digit = len(strnumber)
    
    if shift > digit:
        result = strnumber[::-1]
    else:
        result = strnumber[shift:] + strnumber[:shift]
        
    return int(result)
````

````python
shifting(12345,6)
````

## check password validity

````python
def passwordcheck(password):
    l=0
    u=0
    d=0
    s=0
    for i in password:
        if i.islower():
            l+=1
        if i.isupper():
            u+=1
        if i.isdigit():
            d+=1
        if i in "@_$":
            s+=1
    if l>=1 and u >= 1 and d>=1 and s>=1 and l+u+d+s >= 8:
        print("valid password")
    else:
        print("invalid password")
         
````

````python
passwordcheck('R@1a')
````

## WAP to return the maximum even number from given tuple
- input - (8,7,2,3,4,5,100,3,56)
- outout - 100

````python
def maximum(t):
    max1 = 0
    for i in t:
        if i%2 == 0 and i>max1:
            max1 = i
    print(max1)
````

````python
t = (8,7,2,3,4,5,101,100,3,56)
maximum(t)
````

## PB 314 Ceaser cipher
- input : LJIET ENG
- shift - 3
- Cipher - OMLHW HQJ

````python
def encrypt(message,key):
    message = message.upper()
    alpha = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
    result = ''
    
    for letter in message:
        if letter in alpha:
            letterindex = (alpha.find(letter) + key) % 26
            result = result + alpha[letterindex]
        else:
            result += letter
    return result
````

````python
encrypt("ljiet eyg",3)
````
