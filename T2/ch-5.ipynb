### WAP to find the length of longest consecutive elements in a given list
    input - [6,3,1,2,4]
    output - 4

````python
l = [0,1,2,3,5,7,6,100,8,9,8]
l.sort()
count=max=0
for i in range(len(l)-1):
    if l[i]==l[i+1]-1:
        count+=1
        if count>max:
            max=count
    elif l[i]==l[i+1]:
        continue
    else:
        count =0
print(max+1)
````

````python
def longest(nums):
    result = 0
    nums = set(nums)
    for num in nums:
        if num-1 not in nums:
            current = num
            length = 1
            while current+1 in nums:
                current+=1
                length+=1
            result=max(result,length)
    return result
````

## WAP to create freaquancy count dictionary for words in a given sentamce
    input - the quick brown fox jumps over the lazy dog
    output - {'the':2,'quick':1,..}

````python
s ='the quick brown fox jumps over the lazy dog'
m={}
s = s.split(" ")
for i in s:
    count = 0
    for j in range(0,len(s)):
        if i==s[j]:
            count+=1
    m[i]=count
print(m)
````

````python
inp = 'the quick brown fox jumps over the lazy dog'
freq = {i:inp.count(i) for i in inp.split()}
print(freq)
````
