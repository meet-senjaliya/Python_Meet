# Inheritance
- A class can inherit from other class
- It improves code reuse
- Constructor,attribute and method are inherited to child class
- Parent class has no access to child class
- child class methods overrides the parent class methods
    - this is called mrthod overriding
- ```super()``` method is used to invoke parent class method and constructor

## Single inheritance

````python
class Person:
    def __init__(self,name):
        self.name = name
    def show(self):
        print(self.name + " is a python teacher")
class Professor(Person):
    def display(self):
        print(f"This is {self.name}")
````

````python
per = Professor("Guido Rossum")
````

````python
per.name
````

````python
per.show()
````

````python
per1 =  Person("james")
````

## `super() method`
- if we write conatructor in both parent and child class, the parent class constructor is not available in child class 
- `super()` method is used to call parent class from the child

````python
class Phone:
    def __init__(self,price,brand,camera):
        print("Inside phone class")
        self.price = price
        self.brand = brand
        self.camera = camera
    def buy(self):
        print("Buying phone")
class Smartphone(Phone):
    def buy(self):
        print("Buying a smartphone")
        super().buy()
````

````python
s = Smartphone(20000,"Iphone",13)
````

````python
s.buy()
````

# ABSTRACT CLASS
 - A class derived from ABC class which belongs to abc module is known as `abstruct class` in python
 - ABC class is known as meta class which means class that defines behaviour of other classes
 - Absract class can have concrete and abstract method
 - Abstract class needs to be extends and its methods implemented 
 - One cannot create objects of an abstract class

````python
from abc import ABC, abstractmethod

class DefenceForce(ABC):
    def __init__(self):
        self.id = 101
        
    @abstractmethod
    def area(self):
        pass
    def gun(self):
        print("The Gun is AK 47")
````

````python
class Army(DefenceForce):
    def area(self):
        print("it will throw error",self.id)
````

````python
a = Army()
````

````python
a.area()
````

````python
a.id
````

````python
a.gun()
````

````python
class Navy(DefenceForce):
    def area(self):
        print("Area is lans",self.id)
````

````python
a = Navy()
````

````python
a.area()
````

````python
a.gun()
````

## Create an abstract class shape with abstract method `calculate_area`. create two class Rectangle and circle inheritance from shape.create two objects of rectangle and circle and find the area of these objects

````python
from abc import ABC, abstractmethod
class shape(ABC):
    @abstractmethod
    def calculate_area(self):
        pass
````

````python
class Rectangle(shape):
     def calculate_area(self,l,b):
        self.l=l
        self.b=b
        print("area of rectangle ",self.l*self.b)
class Circle(shape):
    def calculate_area(self,r):
        self.r=r
        print("area of circle ",3.14*self.r**2)
    
````

````python
a = Rectangle()
````

````python
a.calculate_area(1,1)
````

````python
c =Circle()
````

````python
c.calculate_area(2)
````
